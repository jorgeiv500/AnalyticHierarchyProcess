## Primer Caso Embarque de madera

### Uso de matrices y sumaproductos 

 
$\sum_{i=i}^{n}\sum_{j=i}^{n}$ $X_{ij}$ $*$ $C_{ij}$




[<img src="https://github.com/jorgeiv500/diplomado/blob/master/Captura.PNG?raw=true" height="700" width="700">](https://github.com/jorgeiv500/diplomado/blob/master/Captura.PNG?raw=true)



```julia
using JuMP,Cbc, DataFrames

madera=Model(solver=CbcSolver())

costo=[58.5 68.3 47.8 55 63.5;
       65.3 74.8 55 49 57.5;
       59 61.3 63.5 58.8 50]
fuentes=[15;
         20;
         15]
mercados=[11 12 9 10 8]

@variable(madera,x[1:3,1:5]>=0)
@objective(madera,Min,sum(costo.*x))

for i=1:3
@constraint(madera,sum(x[i,j] for j=1:5)==fuentes[i])
end
for j=1:5
@constraint(madera, sum(x[i,j] for i=1:3)==mercados[j])
end
print(madera)
status=solve(madera)
println("el minimo costo es=",getobjectivevalue(madera))
println("las asignaciones son=",DataFrame(getvalue(x)))
```

```julia
using Cbc,JuMP,DataFrames

coeficientes=[2500	2500	2500	2500	2500;
5000	5000	5000	5000	5000;
9000	9000	10000	10000	10000;
18750	18750	25000	25000	25000]

personas=[30	30	30	30	30;
80	80	80	80	80;
200	200	200	200	200;
2000	2000	2000	2000	2000]

servidores=Model(solver=CbcSolver())
@variables servidores begin
    x[1:4,1:5]>=0, Int
end
@objective(servidores,Min,sum(x.*coeficientes))

@constraints servidores begin
sum(personas[1:4,1:2].*x[1:4,1:2])>=60
sum(personas[1:4,1:3].*x[1:4,1:3])>=260
sum(personas[1:4,1:4].*x[1:4,1:4])>=290
sum(personas[1:4,1:5].*x[1:4,1:5])>=365
sum(coeficientes[1:4,1:2].*x[1:4,1:2])<=9500

end
        print(servidores)

status=solve(servidores)
println("el minimo costo es=",getobjectivevalue(servidores))
println("las asignaciones son=",DataFrame(getvalue(x)))
```

```julia
using JuMP,Cbc,DataFrames
c1=[300 300 300;
    10000 10000 10000;
    600 600 600;
    200 200 200;
    0 0 0;
    500 500 500]
c2=[0 0 0;
    400 400 400;
    300 300 300;
    500 500 500;
    10000 10000 10000;
    300 300 300]
c3=[700 700 700;
    500 500 500;
    200 200 200;
    10000 10000 10000;
    400 400 400;
    0 0 0]

nuea=[450;600;550;350;500;450]# numero de estudiantes por area
p6=[0.32;0.37;0.30;0.28;0.39;0.34]
p7=[0.38;0.28;0.32;0.40;0.34;0.28]
p8=[0.30;0.35;0.38;0.32;0.27;0.38]

esexto=nuea.*p6
eseptimo=nuea.*p7
eoctavo=nuea.*p8
capacidad=[900 1100 1000]
estotal=[esexto eseptimo eoctavo]

m=Model(solver=CbcSolver())
@variables m begin
    e1[1:6,1:3]>=0
    e2[1:6,1:3]>=0
    e3[1:6,1:3]>=0
end
@objective(m,Min,sum(e1.*c1)+sum(e2.*c2)+sum(e3.*c3))

for j=1:3

@constraint(m,sum(e1[i,j] for i=1:6)>=0.3*capacidad[1])    
@constraint(m,sum(e1[i,j] for i=1:6)<=0.36*capacidad[1]) 
@constraint(m,sum(e2[i,j] for i=1:6)>=0.3*capacidad[2])    
@constraint(m,sum(e2[i,j] for i=1:6)<=0.36*capacidad[2])    
@constraint(m,sum(e3[i,j] for i=1:6)>=0.3*capacidad[3])    
@constraint(m,sum(e3[i,j] for i=1:6)<=0.36*capacidad[3])    
    
end

for i=1:6
    @constraint(m,e1[i,1]+e2[i,1]+e3[i,1]==esexto[i])
    @constraint(m,e1[i,2]+e2[i,2]+e3[i,2]==eseptimo[i])
    @constraint(m,e1[i,3]+e2[i,3]+e3[i,3]==eoctavo[i])
end
print(m)
status=solve(m)

println("el minimo costo es=",getobjectivevalue(m))
   println("la escuela 1 =",DataFrame(getvalue(e1)))
   println("la escuela 2 =",DataFrame(getvalue(e2)))
println("la escuela 3 =",DataFrame(getvalue(e3)))
```

```julia
using  JuMP, Cbc,DataFrames

proyectos=Model(solver=CbcSolver())
Pij=[100 0 100 267 0;
    400 200 100 153 0;
    200 800 100 99 0;
    200 0 100 451 0;
    100 0 600 30 0]
@variable(proyectos, x[1:5,1:5]>=0 )
@objective(proyectos,Max,sum(Pij.*x))

for i=1:5 #suna de las filas
    @constraint(proyectos, sum(x[i,j] for j=1:5)==1)
end

for j=1:5 #sumatoria de las columnas
    @constraint(proyectos, sum(x[i,j] for i=1:5)==1)
end
print(proyectos)
status=solve(proyectos)
println("el maximo es=",getobjectivevalue(proyectos))
println("las asignaciones son=",DataFrame(getvalue(x)))
```
